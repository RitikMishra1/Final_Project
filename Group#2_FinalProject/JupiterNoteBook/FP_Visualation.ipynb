Job count by Skills Bar Chart

```python

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('HardSkills.csv')
df.head()
skills =list(df['Description'][:10])
jobcnt = list(df['JobCount'][:10])

fig = plt.figure(figsize = (15, 8)) 
# creating the bar plot
plt.bar(skills, jobcnt, color ='lightsalmon',width = 0.4) 
plt.ylabel("No. of Jobs Require the Skill")
plt.title("Hard Skill by Job Count")
plt.show()
```

```python
# Job count by Skills Bar Chart
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('SoftSkills.csv')
df.head()
skills = list(df['Description'])
jobcnt = list(df['JobCount'])
                 
fig = plt.figure(figsize =(10, 7))
plt.pie(jobcnt, labels = skills) 

# show plot
plt.show()
```
